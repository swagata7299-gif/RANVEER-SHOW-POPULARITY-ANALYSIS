```python
# import libraries
import pandas as pd

import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer

df = pd.read_csv(r"C:\Desktop\FDS_Batch_13\Dataset\fds(Ranveer_show).csv")

df
```

```python
# create preprocess_text function
def preprocess_text(text):

    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    # Lemmatize the tokens

    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    # Join the tokens back into a string

    processed_text = ' '.join(lemmatized_tokens)
    return processed_text
# apply the function df

df['Comments'] = df['Comments'].apply(preprocess_text)
df
```

```python

# Initialize the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    # Get the sentiment scores
    scores = analyzer.polarity_scores(text)

    # Determine sentiment based on the scores
    if scores['pos'] > 0.1:
        sentiment = 1  # Positive
    elif scores['neg'] > 0.1:
        sentiment = -1  # Negative
    else:
        sentiment = 0  # Neutral

    return sentiment

df['sentiment'] = df['Comments'].apply(get_sentiment)
# Save the DataFrame with sentiment analysis results to a CSV file
output_file_path = r"C:\Desktop\FDS_Batch_13\Dataset\sentiment_analysis_output.csv"
df.to_csv(output_file_path, index=False)

df

```

```python
# Import necessary module
from sklearn.metrics import confusion_matrix
# Ensure 'Label' and 'sentiment' columns are of type string
df['Label'] = df['Label'].astype(str)
df['sentiment'] = df['sentiment'].astype(str)

# Check unique values in the columns
unique_labels = sorted(df['Label'].unique())
unique_sentiments = sorted(df['sentiment'].unique())

print("Unique labels in 'Label':", unique_labels)
print("Unique labels in 'sentiment':", unique_sentiments)

# Generate confusion matrix
conf_matrix = confusion_matrix(df['Label'], df['sentiment'], labels=unique_labels)

print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.metrics import classification_report

print(classification_report(df['Label'], df['sentiment']))
```
